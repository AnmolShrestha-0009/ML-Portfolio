In this notebook we will see data agumentation and generate diff img using single image


First we will import necessary libary

```python
import numpy as np
import keras
from keras import layers
from keras import Sequential
```

Now we will create a pipeline for agumnentaion


```python
data_agumentation=keras.Sequential([
    layers.RandomFlip("horizontal_and_vertical"),
    layers.RandomRotation(factor=0.8),
    layers.RandomZoom(height_factor=0.5,width_factor=0.5),
    layers.RandomTranslation(height_factor=0.15, width_factor=0.5),
    layers.RandomBrightness(factor=0.1)
])
```

Now we will load image

```python
img_path='dog.png' #img_path='dog.your_image'
img=keras.utils.load_img(img_path)
```

now we will change the image to numpy array. this image consist of rgb channel(255,255,3)

```python
array_img=keras.utils.img_to_array(img)
```

keras expect the image in batches or 4d tensor but we have only 2d tensor so we will change to batches using numpy

```python
batch_input=np.expand_dims(array_img,axis=0)
```

now we fill pass this through our pipelines of data_agumentation

```python
agumented_data=data_agumentation(batch_input,training=True)
```

This is our 4 d tensor now we will clip it to 3d image

```python
output_array_data=np.clip(agumented_data[0],0,255)
```

We again transform our array to image

```python
output_img=keras.utils.array_to_img(output_array_data)
```

Our agumneted output is ready now we will fetch it

```python
output_img.save('a.png')
```

Workflow
create pipeline<br>
load image<br>
convert image to array 4d tensor<br>
convert to batch<br>
send it to pipeline<br>
clip the batch image to 3d tensor<br>
convert toimage<br>
save the image<br>
